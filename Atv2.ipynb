```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

Reading and extracting informations
altura, largura, Canais de cor, tipo de dado, tom de cinza máximo, tom de
cinza médio, tom de cinza mínimo.

```python
def image(path: str):
    img = cv.imread(path)
    h, w, chanelColor = img.shape[0], img.shape[1], img.shape[2]
    print(f'Sua altura é {h}')
    print(f'Sua largura é {w}')
    if chanelColor == 3:
        print(f'Sua imagem tem {chanelColor} canais de cores, podendo ser RGB')
    if chanelColor == 1: 
        print(f'Sua imagem tem {chanelColor} canais de cores, podendo ser em tons de cinza')
    else:
        print(f'Sua imagem tem {chanelColor} canais de cores')

    # Convertendo imagem para tons de cinza
    
    img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    max_gray = np.max(img_gray)
    min_gray = np.min(img_gray)
    mean_gray = np.mean(img_gray)

    print(f'O tom máximo de cinza é {max_gray}\n O tom mínimo de cinza é {min_gray}\n O tom médio de cinza é {mean_gray}')
```

```python
image('cyberpunk.png')
```

```python
def filterImage(path:str):
    img = cv.imread(path)
    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    #applied gaussian, median and mean filter
    gaussianBlur = cv.GaussianBlur(img_rgb, (11,11), 0)
    medianBlur = cv.medianBlur(img_rgb, 11)
    meanBlur = cv.blur(img_rgb, (11,11))

    fig = plt.figure(figsize=(10,7))

    fig.add_subplot(1,3,1)
    plt.imshow(gaussianBlur)
    plt.title('Gaussian Blur')
    fig.add_subplot(1,3,2)
    plt.imshow(medianBlur)
    plt.title('Median Blur')
    fig.add_subplot(1,3,3)
    plt.imshow(meanBlur)
    plt.title('Mean Blur')

    
    return plt.show()
```

```python
filterImage('cyberpunk.png')
```
